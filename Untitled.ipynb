```python
x=t.loc["1985":"2014"]
x
```

```python
aa=AA(x,max_p=9,start_p=1,d=2,start_q=1,max_q=9,max_P=3,max_Q=3,start_P=1,start_Q=1,m=12,trace=True,n_jobs=-1)
```

```python
aa.summary()
```

```python
ypre=aa.predict(n_periods=37)
ypre
```

```python
#mp.plot(x,label="Train")
mp.plot(ypre,label="Predict",linestyle="--",)
mp.plot(y,label="Actual")
mp.xticks(rotation=45)
mp.legend();
```
